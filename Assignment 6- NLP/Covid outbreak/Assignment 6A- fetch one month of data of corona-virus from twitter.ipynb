```python
#import some important libraries

import string
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from collections import Counter
import pandas as pd
import seaborn as sns
from nltk.corpus import stopwords
import GetOldTweets3 as gt

```

```python
#stop words
punt=string.punctuation
stop_words = stopwords.words('english')
print("Stop_words:--->\n",stop_words)
print('\nPUNCTUATIONS:\n',punt)
```

```python
# data of coronavirus from twitter before it's outbreak
       
tweets_before_outbreak = gt.manager.TweetCriteria().setQuerySearch('coronavirus ').setSince("2019-12-07").setUntil("2020-01-06") \
        .setMaxTweets(600)
   
# Creation of list that contains all tweets
before_tweets = gt.manager.TweetManager.getTweets(tweets_before_outbreak)
# Creating list of chosen tweet data
twt_before = [[tweet.text.lower()] for tweet in before_tweets]
```

```python
# reading text file
print("Tweets Found:",len(twt_before))
```

```python
print("Corona virus tweet before outbreak ==>", twt_before)
```

```python
#cleaning  "text_before"
tweets= ""
for i in range(0, len(twt_before)):
    tweets = twt_before[i][0] + " " + tweets

lower = tweets.lower()

# Removing punctuation marks
cleaned_before= lower.translate(str.maketrans('', '', string.punctuation))

# word tokenizarion using NLTK
tokenized_before = word_tokenize(cleaned_before)
```

```python
print("No of words extracted:",len(tokenized_before),"\n Words: \n",tokenized_before)
```

```python

# Removing stop words from the tokenized words list
final_words = [word for word in tokenized_before if word not in stop_words]

filtered_tweets = ""
for i in range(0, len(twt_before)):
    filtered_tweets = twt_before[i][0] + " " + filtered_tweets

lower = filtered_tweets.lower()
# Removing punctuation marks
cleaned_before= lower.translate(str.maketrans('', '', string.punctuation))

# word tokenizarion using NLTK
tokenized_before = word_tokenize(cleaned_before)
        
```

```python
print("No of Tokenized word found:",len(tokenized_before),
     "\nNo of Final Words",len(filtered_tweets),
     "\nNo of words dropped", len(tokenized_before)-len(filtered_tweets))
```

```python

print("\nfinalwords_before:\n",filtered_tweets)
```

```python
tweets_after_outbreak = gt.manager.TweetCriteria().setQuerySearch('covid-19').setSince("2020-01-05").setUntil("2020-02-04") \
        .setMaxTweets(600)

after = gt.manager.TweetManager.getTweets(tweets_after_outbreak)
   
twt_after = [[i.text.lower()] for i in after]
```

```python
print("Tweets Found:",len(twt_after))
```

```python
print("Tweets_Found\n\n",twt_after)
```

```python
#cleaning the texts in "text_after"
filtered_tweets1 = ""
for i in range(0, len(twt_after)):
    filtered_tweets1 = twt_after[i][0] + " " + filtered_tweets1

lower = filtered_tweets1.lower()

# Removing punctuation marks
cleaned_after= lower.translate(str.maketrans('', '', string.punctuation))

# word tokenizarion using NLTK
tokenized_after = word_tokenize(cleaned_after)
```

```python
print("No of words extracted:",len(tokenized_after),"\n Words: \n",tokenized_after)
```

```python
#removing stop words from tokenized_before
finalwords_after_tweets=[i for i in tokenized_after if i not in punt if i not in stop_words]
```

```python
print("No of Tokenized word found:",len(tokenized_after),
     "\nNo of Final Words",len(finalwords_after_tweets),
     "\nNo of words dropped", len(tokenized_after)-len(finalwords_after_tweets))
```

```python

print("\nfinalwords_after:\n",finalwords_after_tweets)
```

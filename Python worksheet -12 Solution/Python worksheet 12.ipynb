## Q13. Now check the percentage of missing values in each column again, if missing values exist, treat it with the most suitable method.

```python
for i in df.columns:
    null=round(((df[i].isnull().sum()/df.shape[0])*100),2)
    if null >0:
        print('percentage of null values in',i,'= ',null)
```

```python
for i in df.columns:
    null=round(((df[i].isnull().sum()/df.shape[0])*100),2)
    if null >0:
        if df[i].dtype == 'float64':
            df[i].fillna(df[i].mean(),inplace=True)
            
        if df[i].dtype == 'object':
            df[i].fillna(df[i].mode()[0],inplace=True)
```

#### 1.  Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from selenium import webdriver
driver=webdriver.Chrome(executable_path=r"C:\Users\SAURABH UPADHYAY\driver\cd.exe")
```

```python
headers=["h1","h2","h3","h4","h5","h6"]  #there are 6 different header tags are in HTML
```

```python
url="https://en.wikipedia.org/wiki/Main_Page"
driver.get(url)
driver.implicitly_wait(10)
```

```python
for i in headers:
    k=0
    head=driver.find_elements_by_tag_name(i)
    if len(head)>0:
        print("\n\nWith tag name ",i,' number of elements found=', len(head),'As dispalyed below\n')
        for h in head:
            print('\t',k,'---',h)
            k=k+1
    else:
        print("\n\nWith tag name ",i,' no element found')
```

```python
driver.quit()
```

#### 2.Write a python program to display IMDB’s Top rated 100 movies’ data(i.e. Name, IMDB rating, duration, director, star cast - 10 main characters)

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait as wdt
from selenium.webdriver.support import expected_conditions as ec
from selenium.common.exceptions import StaleElementReferenceException as ser
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd, time
```

```python
#scrapping news
movie={}
movie['NAME']=[] #for NAME
movie['RATING']=[] #for IMDB RATING
movie['DURATION']=[]  #for MOVIE DURATION
movie['DIRECTOR']=[]  #for MOVIE'S DIRECTOR NAME
movie['STAR_CAST']=[]  #for MOVIE DURATION

link=[]

def scrapper(driver):
       
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")   #TO SCROLL DOWN THE BOTTOM
    time.sleep(12)    
    
   
    l1=driver.find_elements_by_xpath("//td[@class='titleColumn']//a")  #name
    names=l1[0:100]
    
    l2=driver.find_elements_by_xpath("//td[@class='ratingColumn imdbRating']")  #rating
    ratings=l2[0:100]

#for movie name    
    for n in names:
        movie['NAME'].append(n.text)

#for rating        
    for r in ratings:
        movie['RATING'].append(r.text)

##getting link of all names   
    for li in names:
        link.append(li.get_attribute('href'))

```

```python
chrome_options = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values.notifications" : 2}
chrome_options.add_experimental_option("prefs",prefs)
```

```python
driver=webdriver.Chrome(executable_path=r"C:\Users\SAURABH UPADHYAY\driver\cd.exe")
```

```python
url="https://www.imdb.com/"
driver.get(url)
driver.implicitly_wait(12)
driver.maximize_window()
```

```python
driver.find_element_by_xpath("//*[@id='imdbHeader-navDrawerOpen--desktop']").click()
time.sleep(3)
#clicking on drop down button
```

```python
top=driver.find_element_by_xpath("//a[@href='/chart/top/?ref_=nv_mv_250']")
top.click()
time.sleep(7)
#clicking on top rated movie link
```

```python
print('Title:' ,driver.title.upper())
try:
    scrapper(driver=driver)
    time.sleep(7)
except ser:
    print('error while scrapping')
    error_url= driver.current_url
    driver.get(error_url)
    scrapper(driver=driver)
```

```python
%%time
for i in link:
    driver.get(i)
    time.sleep(5)
       
    duration=driver.find_element_by_tag_name('time').text
    movie['DURATION'].append(duration)       #getting duration of each movies
    
    fullcast_button=driver.find_element_by_class_name("quicklink")                  #opening full cats and crew link
    fullcast_button.click()
    
    
    director= driver.find_element_by_xpath("//td[@class='name']//a").text
    movie['DIRECTOR'].append(director)  #directors name
    
    cast=driver.find_elements_by_xpath("//td[@class='primary_photo']//img")   #taaking the list of cast and crew
    casts=cast[0:10]    #taking 10 cast name 
    actor=[]
    for i in casts:
        actor.append(i.get_attribute('alt'))      
    movie['STAR_CAST'].append(actor)
    
    time.sleep(5)
```

```python
print(len(movie['NAME']))
print(len(movie['RATING']))
print(len( movie['STAR_CAST']))
```

```python
top_movies=pd.DataFrame(movie)
```

#### 4.  Write  a  python  program  to  scrap  book  name,  author  name,  genre  and  book  review  of  5  books  from

```python

from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
```

```python
#scrapping news
book={}
book['NAME']=[] #for NAME
book['AUTHOR']=[] #for IMDB RATING
book['GENRE']=[]  #for MOVIE DURATION
book['REVIEW']=[]  #for MOVIE'S DIRECTOR NAME
```

```python
chrome_options = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values.notifications" : 2}
chrome_options.add_experimental_option("prefs",prefs)

driver=webdriver.Chrome(executable_path=r"C:\Users\SAURABH UPADHYAY\driver\cd.exe")
```

```python
url="https://bookpage.com/"
driver.get(url)
driver.implicitly_wait(12)
driver.maximize_window()
```

```python
bookreviews=driver.find_element_by_xpath("//a[@href='/reviews']")

bookreviews.click()
time.sleep(4)
#clicking on review section
```

```python
rev=driver.find_element_by_xpath("//ul[@class='dropdown-menu']//a")
rev.click()
time.sleep(5)
```

```python
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")   #TO SCROLL DOWN THE BOTTOM
time.sleep(12)    

l1=driver.find_elements_by_xpath("//h4//a")  #BOOK name
books=l1[0:5]
    
l2=driver.find_elements_by_xpath("//p[@class='sans bold']")  #AUTHOR
authors=l2[0:5]
    
l3=driver.find_elements_by_xpath("//p[@class='genre-links hidden-phone']")  #GENERE
gen=l3[0:5]

l4=[]
for i in books:
    l4.append(i.get_attribute('href'))  #REVIEW LINK
```

```python
book={}
book['NAME']=[] #for NAME
book['AUTHOR']=[] #for IMDB RATING
book['GENRE']=[]  #for MOVIE DURATION
book['REVIEW']=[]  #for MOVIE'S DIRECTOR NAME
```

```python
for b in books:
    book['NAME'].append(b.text)

for a in authors:
    book['AUTHOR'].append(a.text)

for g in gen:
    book['GENRE'].append(g.text)

for s in l4:
    driver.get(s)
    desc=driver.find_elements_by_xpath("//div[@class='article-body']//p")
   
    article=''
    
    for des in desc:
        article = article + ' ' +  des.text
    book['REVIEW'].append(article)
```

```python
print(len(book['NAME']))
print(len(book['AUTHOR']))
print(len(book['GENRE']))
print(len(book['REVIEW']))
```

```python
Book=pd.DataFrame(book)
```
